### Task1

Найти похожие слова по триграмам.

```python
from nltk.book import *

def get_etalon(b):
    return [b[i:i+n] for i in range(len(b)-n+1)]

b = "university"
n = 3
etalon = get_etalon(b)
etalon
```

```python
all = {}
for word in text8:
    trigrams = get_etalon(word)
    if trigrams != []:
        all[word] = trigrams
all
```

```python
def intersection(lst1, lst2): 
    return [value for value in lst1 if value in lst2] 

intersections = {}
for key, value in all.items():
    inters = intersection(value, etalon)
    if inters != []:
        intersections[key] = len(inters)

sorted(intersections.items(), key = lambda kv: kv[1], reverse = True)    
```
